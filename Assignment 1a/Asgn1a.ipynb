```python
#Pseudo Random Number Generator
def pRNG(seed, c, num):
    x = seed
    L = []
    itr = 0 
    while itr < num:
        x = c*x*(1 - x)
        itr += 1
        L.append(x)
    return L

#Generating randdom numbers list from pRNG function defined above
RN1 = pRNG(0.1, 3.9986452312, 500)
RN2 = pRNG(0.1, 3.9976543214, 500)
RN3 = pRNG(0.1, 3.99864460122, 500)
RN4 = pRNG(0.1, 3.98864460122, 500)
RN5 = pRNG(0.1, 3, 1000)

print('The correlation test score for first set of paramters is:',correlation_test(5,RN1))
print('The correlation test score for second set of paramters is:',correlation_test(5,RN2))
print('The correlation test score for third set of paramters is:',correlation_test(5,RN3))
print('The correlation test score for fourth set of paramters is:',correlation_test(5,RN4))
print('The correlation test score for fifth set of paramters is:',correlation_test(5,RN5))

plt.plot(RN1)
plt.show()
plt.plot(RN2)
plt.show()
plt.plot(RN3)
plt.show()
plt.plot(RN4)
plt.show()
plt.plot(RN5)
plt.show()
coplot(RN1,5)
coplot(RN1,7)
coplot(RN1,10)
coplot(RN2,5)
coplot(RN3,5)
coplot(RN4,5)
coplot(RN5,5)

#Output
'''
The correlation test score for first set of paramters is: 0.006530360360857024
The correlation test score for second set of paramters is: -0.007805636648806458
The correlation test score for third set of paramters is: 0.0018188069001494056
The correlation test score for fourth set of paramters is: -0.00013402166958997874
The correlation test score for fifth set of paramters is: -0.0021388103291917204
'''
```
